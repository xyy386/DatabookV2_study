```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
A = imread(os.path.join('dog.jpg'))
X = np.mean(A,-1)
img = plt.imshow(X)
img.set_cmap('gray')
```

```python
U, Sigma, V_T = np.linalg.svd(X, full_matrices=False)
Sigma = np.diag(Sigma)
```

```python
for r in (5,20,100):
    X_approx = U[:,:r] @ Sigma[0:r,:r] @V_T[:r,:]
    img = plt.imshow(X_approx)
    img.set_cmap('gray')
    plt.show()
```

```python
plt.semilogy(np.diag(Sigma))
plt.show(np.cumsum(np.diag(Sigma))/np.sum(np.diag(Sigma)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

Code1.13: Creating tensor data

```python
x = np.arange(-5, 5.01, 0.1)
y = np.arange(-6, 6.01, 0.1)
t = np.arange(0, 10 * np.pi + 0.01, 0.1)
```

```python
X, Y, T = np.meshgrid(x,y,t)
```

```python
A = np.exp(-(X ** 2 + 0.5 * Y ** 2)) * np.cos(2 * T) + (1 / np.cosh(X)) * np.tanh(X) * np.exp(-0.2 * Y ** 2) * np.sin(T)
```

Code1.14: Two_factor tensor model

```python
from tensorly.decomposition import parafac
```

```python
weights, factors = parafac(A, rank=2)
A1, A2, A3 = factors

print("weights shape:", weights.shape)
print("A1 shape:", A1.shape)
print("A2 shape:", A2.shape)
print("A3 shape:", A3.shape)
```

```python
fig, axs = plt.subplots(3,1)
axs[0].plot(y,A1,linewidth=2)
axs[1].plot(x,A2,linewidth=2)
axs[2].plot(t,A3,linewidth=2)
plt.show()
```

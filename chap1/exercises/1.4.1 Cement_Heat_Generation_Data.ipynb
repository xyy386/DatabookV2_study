```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
A = np.loadtxt('../../DATA/hald_ingredients.csv', delimiter=',')
A
```

```python
b = np.loadtxt('../../DATA/hald_heat.csv', delimiter=',')
b
```

```python
U, Sigma, V_T = np.linalg.svd(A,full_matrices=False)
x = V_T.T @ np.linalg.inv(np.diag(Sigma)) @ U.T @ b
x
```

```python
plt.plot(b, color='black',label='Heat Data')
plt.plot(A@x,color='red',label='Regression')
plt.legend(loc='best')
plt.show()
```

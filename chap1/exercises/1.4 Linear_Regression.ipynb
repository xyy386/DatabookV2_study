```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 3
a = np.arange(-2,2,0.25)
a
```

```python
a = a.reshape(-1,1)
a
```

```python
# 噪声
b = x * a + np.random.randn(a.shape[0],a.shape[1])
b
```

```python
# 可视化
plt.plot(a,b,'x',color='red',markersize=10,label='noisy data')
plt.legend(loc='best')
plt.show()
```

```python
# 计算SVD
U, Sigma, V_T = np.linalg.svd(a, full_matrices=False)
```

```python
# 最小二乘法
x_tilde = V_T.T @ np.linalg.inv(np.diag(Sigma)) @ U.T @ b
x_tilde
```

```python
# 计算拟合
b_tilde = x_tilde * a
```

```python
# 可视化
plt.plot(a,b_tilde,'--',label='Regression line')
plt.show()
```

```python
# 画在一个图中
plt.plot(a,b,'x',color='red',markersize=10,label='noisy data')
plt.plot(a,b_tilde,'--',label='Regression line')
plt.plot(a,a*x,'-',color='black',markersize=10,label='Ture data')
plt.legend(loc='best')
plt.show()
```

```python
# 另外一种求x的方法
x_tilde_2 = np.linalg.pinv(a) @ b
x_tilde_2
```

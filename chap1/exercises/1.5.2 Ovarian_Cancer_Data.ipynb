```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
obs = np.loadtxt('../../DATA/ovariancancer_obs.csv', delimiter=',')
f = open('../../DATA/ovariancancer_grp.csv', 'r')
grp = f.read().split('\n')
```

```python
# 进行SVD分解
U, Sigma, V_T = np.linalg.svd(obs,full_matrices=False)
```

```python
plt.plot(Sigma,'-o',color = 'black',markersize = 5,markerfacecolor='none')
plt.show()
```

```python
sig_value_sum_rate = []
r_array = np.arange(0,len(Sigma),1)
for r in r_array:
    sig_value_sum_rate.append(np.sum(Sigma[:r+1])/np.sum(Sigma))
plt.plot(sig_value_sum_rate,'-o',color = 'black',markersize = 5,markerfacecolor='none')
plt.xlim(0-1,len(Sigma)+1)
plt.ylim(0.5,1)
plt.show()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
for j in range(obs.shape[0]):
    x = V_T[0,:] @ obs[j,:].T
    y = V_T[1,:] @ obs[j,:].T
    z = V_T[2,:] @ obs[j,:].T
    
    if grp[j] == 'Cancer':
        ax.scatter(x,y,z,marker='x', color='red', s=20)
    else:
        ax.scatter(x,y,z,marker='o', color='blue', s=20)
        

ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
plt.show()
```

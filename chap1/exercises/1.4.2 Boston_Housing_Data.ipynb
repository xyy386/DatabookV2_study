```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = np.loadtxt('../../Data/housing.data')
H
```

```python
A = H[:,0:13]
A
```

```python
b = H[:,-1]
b
```

```python
A = np.hstack((A, np.ones((A.shape[0],1))))
A
```

```python
U, Sigma, V_T = np.linalg.svd(A,full_matrices=False)
```

```python
x = V_T.T @ np.linalg.inv(np.diag(Sigma)) @ U.T @ b
x
```

```python
plt.plot(b,color='black',label = 'Housing value')
plt.plot(A@x ,color='red',label = 'Regression')
plt.legend(loc='best')
plt.show()
```

```python
sort_idx = np.argsort(b)
sort_idx
```

```python
b_sort = b[sort_idx]
b_sort
```

```python
result_sort = (A @ x)[sort_idx]
result_sort
```

```python
plt.plot(b_sort,color='black',linewidth=2,label = 'Housing value')
plt.plot(result_sort,'-o',color='red',linewidth=1, markersize=2,label = 'Regression')
plt.legend(loc='best')
plt.show()
```
